The dataset contains detailed information on 700 mobile device users, capturing various usage patterns and behavior classifications. The features include app usage time, screen-on time, battery drain, data consumption, and more. In addition, user demographics like age and gender are included, as well as the device model and operating system. The 'user behavior class' attribute categorizes users based on their usage patterns, ranging from light to extreme behavior. All features, except for the 'user behavior class', can be used as input variables for analysis and modeling, while the 'user behavior class' serves as the target variable for prediction. ## Handling the Missing Values
Handle the missing values using suitable method based on the insights obtained from the various visualizations.

```python
# Write your code here. Add as many boxes as you need.
df = df.dropna(subset=['Age'],axis=0)
```

```python
df = df.fillna(df['Data Usage (MB/day)'].median())
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy= 'most_frequent')
```

```python
df['Operating System'] = df['Operating System'].astype(str)
```

```python
df['Operating System'] = imputer.fit_transform(df['Operating System'].values.reshape(1, -1))[0]
```

```python
df.isnull().sum()
```

```python
df = df.drop('User ID',axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df[['App Usage Time (min/day)',	'Screen On Time (hours/day)', 'Battery Drain (mAh/day)', 'Number of Apps Installed', 'Data Usage (MB/day)', 'Age']] = scaler.fit_transform(df[['App Usage Time (min/day)',	'Screen On Time (hours/day)', 'Battery Drain (mAh/day)', 'Number of Apps Installed', 'Data Usage (MB/day)', 'Age']])
```

```python
df['Operating System'] = encoder1.fit_transform(df['Operating System'])
```

## Dataset Splitting
Partition the dataset into training and testing sets with an 80:20 ratio.

```python
# Write your code here. Add as many boxes as you need.
X = df.drop(columns=['User Behavior Class', 'Age', 'Operating System'])
Y = df['User Behavior Class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```
