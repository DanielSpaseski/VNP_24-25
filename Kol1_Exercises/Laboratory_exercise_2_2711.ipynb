## Handling the Missing Values
Handle the missing values using suitable method based on the insights obtained from the various visualizations.

```python
# Write your code here. Add as many boxes as you need.
df = df.drop(columns=['id', 'Unnamed: 0'])
```

```python
num_cols = ['Age', 'Height', 'Weight'] 
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
cat_cols = ['Gender', 'SMOKE', 'MTRANS', 'NObeyesdad']
frequent_imputer = SimpleImputer(strategy = 'most_frequent')
df[cat_cols] = frequent_imputer.fit_transform(df[cat_cols])
```

```python
def knn_impute_data(data:pd.DataFrame, columns:list, n_neighbors:int):
   imputer = KNNImputer(n_neighbors=n_neighbors)
   data_copy = data.copy()

   for column in columns:
     data_copy[column] = imputer.fit_transform(data_copy[column].to_numpy().reshape(-1, 1))
   return data_copy
```

```python
df = knn_impute_data(data=df, columns=['FCVC', 'NCP', 'CH2O', 'FAF', 'TUE'], n_neighbors=5)
```

```python
df.isnull().sum()
```

## Feature Preprocessing


Preprocess the features to make them ready for using the model

```python
# Write your code here. Add as many boxes as you need.
encoder = LabelEncoder()
df['Gender'] = encoder.fit_transform(df['Gender'])
df['MTRANS'] = encoder.fit_transform(df['MTRANS'])
df['SMOKE'] = encoder.fit_transform(df['SMOKE'])
df['NObeyesdad'] = encoder.fit_transform(df['NObeyesdad'])
```

## Hyper-parameter optimization

Train an XGBoost Model, optimize it's hyper-parameters with grid-search and 5-fold cross validation on the training set.

```python
# Write your code here. Add as many boxes as you need.
X = df.drop(columns=['FAF'],axis=1)
Y = df['FAF']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
model = XGBRegressor()
```

```python
param_grid = {
    'n_estimators': [100, 200,300],
    'learning_rate': [0.01, 0.05,0.1, 0.2],
    'max_depth': [3, 5, 7]
}
```

```python
cv = GridSearchCV(estimator=model,param_grid=param_grid,cv=5,scoring='r2',verbose=1,n_jobs=-1)
```

```python
cv.fit(X_train,Y_train)
cv.best_params_
```

```python
best_model = cv.best_estimator_
Y_pred = best_model.predict(X_test)
```

```python
r2_score(Y_test,Y_pred)
```

# Cross-validation evaluation

Perform leave-one out cross validation on the entire dataset, with an untrained model with the chosen hyper-parameters from the previous step.

```python
# Write your code here. Add as many boxes as you need.
scores = []
kfold = KFold(n_splits=5)
for train_index, test_index in kfold.split(X):
    X_train, Y_train = X.values[train_index], Y.values[train_index]
    X_test, Y_test = X.values[test_index], Y.values[test_index]
    
    model = XGBRegressor(n_jobs=-1,**cv.best_params_)
    model.fit(X_train, Y_train)
    Y_pred = model.predict(X_test)
    scores.append(mean_squared_error(Y_test, Y_pred))
print(f"K Average mse_scores: {np.mean(scores)}")
```
